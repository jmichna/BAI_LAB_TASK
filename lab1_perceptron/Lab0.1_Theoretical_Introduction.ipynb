## 1. Co to jest Perceptron?

Wyobraź sobie najprostszą maszynę decyzyjną. Perceptron działa jak funkcja, która przyjmuje zestaw danych (np. informacje o pogodzie), przemnoży je przez ich "ważność" i wypluje jedną odpowiedź: **TAK (1)** lub **NIE (0)**.

W kontekście uczenia maszynowego, perceptron jest **klasyfikatorem liniowym**. Oznacza to, że próbuje on rozdzielić dane na dwie grupy za pomocą linii prostej (w 2D), płaszczyzny (w 3D) lub hiperpłaszczyzny (w wielu wymiarach).

### Architektura modelu:
1. **Wejścia ($x_1, x_2, ..., x_n$):** To są Twoje dane. Jeśli trenujesz model na obrazkach MNIST, każde wejście to jasność jednego piksela.
2. **Wagi ($w_1, w_2, ..., w_n$):** Najważniejszy element. Każde wejście ma swoją wagę. Jeśli piksel w centrum obrazka częściej występuje w cyfrze "1", jego waga będzie rosła.
3. **Bias ($b$):** To stała wartość dodawana do sumy. Pozwala ona modelowi przesunąć linię decyzyjną tak, aby nie musiała ona przechodzić przez punkt (0,0).

## 2. Dlaczego używamy NumPy zamiast pętli `for`?

To jest kluczowy moment dla zrozumienia wydajności w Pythonie. W klasycznym programowaniu (np. pisząc kółko i krzyżyk) może Ci się wydawać, że pętla `for` przejrzeń listę to prosta sprawa. Jednak w uczeniu maszynowym operujemy na tysiącach (lub milionach) danych.

### Problem z pętlą `for` w Pythonie:
Python jest językiem interpretowanym. Kiedy piszesz:
```python
suma = 0
for i in range(len(x)):
    suma += x[i] * w[i]
```
Interpreter Pythona przy każdej iteracji musi:
- Sprawdzić typ zmiennej `x[i]`.
- Sprawdzić typ zmiennej `w[i]`.
- Pobrać odpowiednią funkcję mnożenia dla tych typów.
- Wykonać operację i zapisać wynik.

To powoduje ogromny narzut czasowy (overhead). Dla wektora o długości 784 pikseli (MNIST), wykonujesz 784 spowolnione operacje.

### Potęga `np.dot` (Iloczyn skalarany):
`np.dot(x, w)` to operacja **wektoryzowana**. 
1. **Implementacja w C/Fortranie:** NumPy pod spodem używa skompilowanego kodu, który działa bezpośrednio na pamięci procesora.
2. **SIMD (Single Instruction, Multiple Data):** Nowoczesne procesory potrafią pomnożyć cały blok liczb (np. 8 lub 16 naraz) za pomocą jednej instrukcji sprzętowej. Pętla `for` tego nie potrafi.
3. **Lokalność pamięci:** NumPy trzyma dane w ciągłych blokach pamięci, co pozwala procesorowi błyskawicznie ładować je do pamięci podręcznej (cache).

Poniżej zobaczysz porównanie w kodzie.

````python
import numpy as np
import time

# Tworzymy duże wektory danych (np. 1 000 000 elementów)
size = 1000000
x = np.random.rand(size)
w = np.random.rand(size)

# 1. Obliczanie za pomocą pętli for
start_time = time.time()
suma_for = 0
for i in range(size):
    suma_for += x[i] * w[i]
time_for = time.time() - start_time

# 2. Obliczanie za pomocą np.dot
start_time = time.time()
suma_dot = np.dot(x, w)
time_dot = time.time() - start_time

print(f"Czas pętli for: {time_for:.5f} sek")
print(f"Czas np.dot:     {time_dot:.5f} sek")
print(f"NumPy jest {time_for / time_dot:.1f} razy szybszy!")
````

## 3. Matematyka Perceptronu

Model działa w dwóch krokach:

### Krok 1: Suma ważona (z)
$$z = (x_1 \cdot w_1 + x_2 \cdot w_2 + ... + x_n \cdot w_n) + b$$
W NumPy zapisujemy to błyskawicznie:
```python
z = np.dot(x, w) + b
```

### Krok 2: Funkcja aktywacji
W perceptronie używamy funkcji progowej. Wynik $z$ zamieniamy na 0 lub 1.
$$\hat{y} = 1 \text{ jeśli } z \ge 0 \text{, w przeciwnym razie } 0$$

## 4. Algorytm Uczenia

Perceptron uczy się na błędach. Jeśli poda zły wynik, korygujemy wagi.

1. **Oblicz błąd:** 
   `error = spodziewany_wynik - wynik_perceptronu`
   (Wartość `error` może wynosić 1, -1 lub 0 jeśli trafiłeś).

2. **Aktualizuj wagi:**
   `w_nowe = w_stare + eta * error * x` 
   *`eta` (learning rate) określa jak duże kroki wykonujemy.*

3. **Aktualizuj bias:**
   `b_nowy = b_stary + eta * error`

Operacja `eta * error * x` w NumPy jest również błyskawiczna, ponieważ mnożymy liczbę przez cały wektor naraz.
